```python
import numpy as np
import pandas as pd
```

```python
# The make_table function is used to create a new DataFrame that
# consists of information that we need. There is a lot of information of all the other teams as well as
# Betting odds, which is really not necessary for this project. 
# This function only filters out the necessary information
# i.e. Match Details (data variable) and Arsenal's individual match detail.

def make_table(data):
    data = data[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC', 'HF', 'AF', 'HY', 'AY', 'HR', 'AR']]
    away_data = data[data['AwayTeam'] == 'Arsenal']
    home_data = data[data['HomeTeam'] == 'Arsenal']
    final_data = pd.concat([away_data, home_data])
    return final_data
```

```python
# Reading each CSVs

data1718 = pd.read_csv("./data/season-1718_csv.csv")
data1617 = pd.read_csv("./data/season-1617_csv.csv")
data1516 = pd.read_csv("./data/season-1516_csv.csv")
data1415 = pd.read_csv("./data/season-1415_csv.csv")
data1314 = pd.read_csv("./data/season-1314_csv.csv")
data1213 = pd.read_csv("./data/season-1213_csv.csv")
data1112 = pd.read_csv("./data/season-1112_csv.csv")
data1011 = pd.read_csv("./data/season-1011_csv.csv")
data0910 = pd.read_csv("./data/season-0910_csv.csv")
```

```python
# Create a dataframe by calling the function make_data

Arsenal_data = make_table(data1718)
```

```python
# Overwriting the Arsenal Dataframe by calling the make_table function
# and concatenating it with the existing table.

Arsenal_data = pd.concat([Arsenal_data, make_table(data1617)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1516)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1415)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1314)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1213)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1112)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data1011)])
Arsenal_data = pd.concat([Arsenal_data, make_table(data0910)])
```
